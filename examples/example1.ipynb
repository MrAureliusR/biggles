```python
import biggles
import numpy
import math

x = numpy.arange(0, 3 * math.pi, math.pi / 30)
c = numpy.cos(x)
s = numpy.sin(x)

p = biggles.FramedPlot()
p.title = "title"
p.xlabel = r"$x$"
p.ylabel = r"$\Theta$"

p.add(biggles.FillBetween(x, c, x, s))
p.add(biggles.Curve(x, c, color="red"))
p.add(biggles.Curve(x, s, color="blue"))
p
```
