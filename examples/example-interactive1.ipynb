```python
# Using a convenience function to do some quick
# interactive plotting

import numpy
from biggles import plot

x=numpy.linspace(0, 4*numpy.pi,50)
y=numpy.cos(x)*numpy.exp(-x/3)
noise=0.02
yerr = noise + numpy.zeros(x.size)
y += numpy.random.normal(scale=noise, size=y.size)

plot(x, y, yerr=yerr, color='firebrick')
```
